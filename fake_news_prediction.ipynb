## importing dependencies

```python
import numpy as np
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

import nltk
nltk.download('stopwords')
```

##Data pre  processing

### Loading and Labeling Data from Two Files

```python
# Load the first dataset (e.g., representing class 0)
# Assuming the text content is in a column named 'text' and it's a CSV file.
data_class_0 = pd.read_csv('/content/fake.csv') # Replace with your first file path
data_class_0['label'] = 0 # Assign label 0

# Load the second dataset (e.g., representing class 1)
data_class_1 = pd.read_csv('/content/true.csv') # Replace with your second file path
data_class_1['label'] = 1 # Assign label 1

# Combine the two datasets
df = pd.concat([data_class_0, data_class_1], ignore_index=True)

# Display the first few rows of the combined DataFrame to verify
display(df.head())
```

```python
#loading the data to pandas df

df.shape
```

```python
#counting the number of missing values in each column
df.isnull().sum()
```

```python
x=df.drop(columns='label',axis=1)
y=df['label']
```

```python
#stemming the news headlines
port_stem=PorterStemmer()
```

```python
def stemming(content):
  stemmed_content=re.sub('[^a-zA-Z]',' ',content)
  stemmed_content=stemmed_content.lower()
  stemmed_content=stemmed_content.split()
  stemmed_content=[port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content=' '.join(stemmed_content)
  return stemmed_content
```

```python
df['title']=df['title'].apply(stemming)
```

```python
df.head()
```

```python
x=df['title'].values
y=df['label'].values
```

```python
#converting textual data to vectorised data
vectoriser=TfidfVectorizer()
vectoriser.fit(x)
x=vectoriser.transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,shuffle=True,stratify=y,random_state=2)
```
